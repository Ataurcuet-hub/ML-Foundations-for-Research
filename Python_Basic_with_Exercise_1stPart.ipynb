***Smart Way*** **H6: Create a simple calculator that performs basic arithmetic operations (addition, subtraction, multiplication, division) based on user input**

```python

#Smart Calculator
print("press '=' to start and get result finally, c for clear and q for quit")
expression= ""

while True:
  user_input= input("Enter your value: ")
  if user_input == "q":
    print("calculator closed")
    break
  elif user_input== "c":
    expression=""
    print("cleared")

  elif user_input== "=":
   try:
    result= eval(expression)
    print("result:", result)
    print("final expression: ", str(expression))
   except:
    print("invalid expression")
    expression=""
  else:
    expression=expression+user_input
    print("expression: ", expression)

print("Thanks to use my python Calculator")
```

**H8: Create a program that converts temperatures between Fahrenheit and Celsius based on user input.**

```python
## H8: Create a program that converts temperatures between Fahrenheit and Celsius based on user input.
print("Temperature converstion calculator")
print("type c for celsius to farenheit")
print("type f for farenheit to celsius")
print("type q for quit")
while True:
  temp=str(input("c f or q?: "))
  if temp.lower()== "q":
    print("quit the program")
    break
  elif temp.lower()=="f":
    try:
      F=float(input("how much is the temperation in Farenheit: "))
      C=(5/9)*(F-32)
      print(f"Conversion: The tempareture is {C:.2f} degree Celsius")
    except:
      print("invalid input! please type again")
  elif temp.lower()=="c":
    try:
      C=float(input("how much is the temperation in Celsium: "))
      F=(9/5)*C+32
      print(f" COnversion: The temperatur is {F:.2f} degree farenheit")
    except:
      print("invalid input! please type again")
  else:
    print("invalid input! please type 'c', 'f', or 'q'")

print("Thank you!")
```

**# H11: Write a program that prompts the user to input a number N and calculates the sum of all numbers from 1 to N that are divisible by 3 or 5.**

```python
# H11: Write a program that prompts the user to input a number N and calculates the sum of all numbers from 1 to N that are divisible by 3 or 5.
sum=0
N=int(input("N: "))
for x in range(1, N+1):
  if x%3==0 or x%5==0:
    print(x)
    sum=sum+x
    print(f" sum is {sum}")
print(f"final sum of all that numbers is {sum}")

```

**Write a program that prompts the user to input a number n and prints the multiplication table of n (1 to 10) using a for loop.**

```python
# H12: Write a program that prompts the user to input a number n and prints the multiplication table of n (1 to 10) using a for loop.
N=int(input("N: "))
for x in range(2, N+1):
  print(f"well done! Multiplication of {x}")
  for y in range(1, 11):
    multi=x*y
    print(f" {x}*{y}={multi}")
```
