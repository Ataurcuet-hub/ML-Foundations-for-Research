Array **Operation**

````python
a= np.array([1,2,3,4,5])
b=np.array([10,20,30,40,50])
````

````python
print("add: ", a+b)
print("Multi: ", a*b)
print("Power: ", a**2)
print("sqrt: ", np.sqrt(b))
print(a*np.pi)
````

*Statistical Operation* **Numpy**

````python
a= np.array([1,2,3,4,5])
b=np.array([10,20,30,40,50])
print ("Mean:", np.mean(a))
print("standard_devi:" , np.std(a))
print("min:", np.min(b))
print("Max:", np.max(b))
````

````python

fig, axes=plt.subplots(2,2, figsize=(10,8))

#Scatter Plot
#axes[0, 0].scatter(X, Y)
axes[0,0].scatter(df2["math score"][:20], df2["reading score"][:20])
axes[0,0].set_title("scores")

# ১ম সারি, ২য় কলাম (উপরে ডানে)
#axes[0, 1].plot(x_values, y_values)
#df2["gender"].value_counts().plot(kind="bar", ax=axes[0,1])
gender_count=df2["gender"].value_counts()
axes[0,1].bar(gender_count.index, gender_count.values)
axes[0,1].set_title("gender")


# ২য় সারি, ১ম কলাম (নিচে বামে)
axes[1, 0].hist(df2["reading score"], bins=15)
axes[1, 0].set_title("reading score")

# ২য় সারি, ২য় কলাম (নিচে ডানে)
axes[1,1].plot(df2["writing score"] [:30])
axes[1,1].set_title("writing scores")

plt.tight_layout()
plt.show()
fig.savefig("/content/sample_data/multiplot.jpeg", dpi=1200)
````
