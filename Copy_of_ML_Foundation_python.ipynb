Array **Operation**

````python
a= np.array([1,2,3,4,5])
b=np.array([10,20,30,40,50])
````

````python
print("add: ", a+b)
print("Multi: ", a*b)
print("Power: ", a**2)
print("sqrt: ", np.sqrt(b))
print(a*np.pi)
````

*Statistical Operation* **Numpy**

````python
a= np.array([1,2,3,4,5])
b=np.array([10,20,30,40,50])
print ("Mean:", np.mean(a))
print("standard_devi:" , np.std(a))
print("min:", np.min(b))
print("Max:", np.max(b))
````
