```python
import seaborn as sns
import numpy as np
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path= "/content/drive/MyDrive/ML Dataset/StudentsPerformance.csv"
stu=pd.read_csv(file_path)
print(stu.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(stu["math score"])
plt.title("math score plot")
#plt.show()
```

```python
sns.histplot(stu["reading score"], kde= True) #KDE:kernel density estimation
plt.title("reading scores")
plt.show()
```

```python
sns.boxplot(data=stu, x="gender", y="math score")
#sns.boxplot(stu["gender"], stu["math score"])
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.title ("math score by gender")
plt.show()
```

```python
sns.countplot(data=stu, x="race/ethnicity")
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.title("ethinicity group count")
plt.show()
stu["race/ethnicity"].info()
stu["race/ethnicity"].describe()
```

```python
#stu["race/ethnicity"].value_counts().to_frame(name="count")
stu["race/ethnicity"].value_counts()
```

```python
sns.countplot(data=stu, x="gender")
stu["gender"].value_counts()
```

```python
sns.barplot(data=stu, x="gender", y="writing score")
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.7)
plt.title("barplot of writing score by gender")
plt.show()

```

```python
sns.barplot(data=stu, x="parental level of education", y="math score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#kernel density estimation plot
#sns.kdeplot(data=stu, x="reading score", hue="level of education")
sns.kdeplot(data=stu, x="math score", hue="test preparation course")
plt.grid(True, which="both", linestyle="--", linewidth=0.7, alpha=0.6)
plt.savefig("/content/ML out images/kdeplot.jpeg", dpi=600)
plt.show()
```

```python
sns.kdeplot(data=stu, x="reading score", hue="parental level of education")
```

***heatmap***

```python
scores=stu[["math score", "reading score", "writing score"]]
correlation=scores.corr()
print(correlation)
```

```python
sns.heatmap(correlation, annot= True)
plt.title("score correlation")
plt.savefig("/content/ML out images/scoreHeatmap.jpeg", dpi=600)
plt.show()
```

```python
score_data=stu[["math score", "reading score", "writing score"]]
sns.pairplot(score_data)
plt.show()
```

```python
score_gender=stu[["math score", "reading score", "writing score", "gender"]]
sns.pairplot(score_gender, hue="gender")
plt.title("genderwise scores")
plt.savefig("/content/ML out images/scorebygenderPairplot.jpeg", dpi=600)
plt.show()
```
