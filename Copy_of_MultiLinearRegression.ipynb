```python
import pandas as pd
dataset= pd.read_csv('sample_data/USA_Housing.csv')
dataset.head()
```

```python
#data divided to x y
x= dataset[['Avg. Area Income',  'Avg. Area House Age',  'Avg. Area Number of Rooms', 'Avg. Area Number of Bedrooms',  'Area Population']].values
y= dataset[['Price']].values
print(y)
```

```python
#n missing value and no need encoding
# Sacling
from sklearn.preprocessing import StandardScaler
ScX= StandardScaler()
x= ScX.fit_transform(x)
print(x)
```

```python
#Split for train & test
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
#Multiple Linear Regression
from sklearn.linear_model import LinearRegression
MultiRegressor= LinearRegression()
MultiRegressor.fit(xTrain, yTrain)
yPred= MultiRegressor.predict(xTest)

#model good or bad
from sklearn.metrics import r2_score
r2= r2_score(yTest, yPred)
print("r squared value:" +str(r2))
```

```python
#### Polynomial Regression
from sklearn.preprocessing import PolynomialFeatures
PolyFeatures=PolynomialFeatures(degree=2)
xPolyTrain= PolyFeatures.fit_transform(xTrain)
xPolyTest= PolyFeatures.fit_transform(xTest)

from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(xPolyTrain, yTrain)
yPred= regressor.predict(xPolyTest)

r2Poly= r2_score(yTest, yPred)
print( "polinomial prediction:" +str(r2Poly))
```

```python
#spport vector Regression (kernel='rbf')

from sklearn.svm import SVR
svrRegressor= SVR(kernel='rbf')
svrRegressor.fit( xTrain, yTrain )
yPredsvr= svrRegressor.predict(xTest)
r2svr= r2_score(yTest, yPredsvr)
print( "support vector prediction:" +str(r2svr))

```

```python
#Decisio Tree Regressiom
from sklearn.tree import DecisionTreeRegressor
regressorDT=DecisionTreeRegressor()
regressorDT.fit(xTrain,yTrain)
yPredDT=regressorDT.predict(xTest)
r2= r2_score(yPredDT,yTest)
print( "DecisionTree predictio:" +str(r2))
```

```python
#Random Forest Regression
from sklearn.ensemble import RandomForestRegressor
regressorRF=RandomForestRegressor(n_estimators=100)
regressorRF.fit(xTrain, yTrain.ravel())
yPredRF= regressorRF.predict(xTest)
r2RF=r2_score(yTest, yPredRF)
print("Randor Forest Prediction: "+ str(r2RF))

```
